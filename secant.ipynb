```python
import matplotlib.pyplot as plt
def y_equals_f(x):
    return (x) ** 2
x = []
y = []
for i in range(9):
    x.append(i)
    y.append(y_equals_f(i))
for i in range(8, 1, -1):
    plt.plot(x,y, color='blue', linewidth=4)
    plt.plot([x[2], x[i]],[y[2], y[i]], color='red', linewidth=4)
    plt.xticks(size=16, weight='bold')
    plt.yticks(size=16, weight='bold')
    plt.title('Input / Output', weight='bold', size=18)
    plt.xlabel('x', size=16, weight='bold')
    plt.ylabel('y = f(x)', size=16, weight='bold')
    plt.show()
    print('\n')
```
